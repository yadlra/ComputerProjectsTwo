```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = 'images.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Count the number of images per category
category_counts = df['Category'].value_counts()

# Creating the bar chart
plt.figure(figsize=(8, 5))
category_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel('Category')
plt.ylabel('Number of Images')
plt.title('Distribution of Images by Category')
plt.xticks(rotation=0)  # Keep the category names horizontal
plt.show()

```

# Creating a Photo Grid in Python



```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import pandas as pd

# Read the CSV file
df = pd.read_csv('images.csv')

# Filter the DataFrame for a specific category if needed, or use the entire DataFrame
category = 'flowers'  # Choose the category you want to display
filtered_df = df[df['Category'] == category]

# Setup the plot
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))  # Adjust the grid size according to your needs
axs = axs.flatten()

# Load and display the images
for ax, (_, row) in zip(axs, filtered_df.iterrows()):
    img = mpimg.imread(row['Image URL'])  # Make sure the path is accessible
    ax.imshow(img)
    ax.axis('off')  # Hide grid lines
    ax.set_title(row['Caption'])

plt.tight_layout()
plt.show()
```
